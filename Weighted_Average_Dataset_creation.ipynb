```python
from math import radians, sin, cos, asin, sqrt
def cal_distance(lat1, lat2, lon1, lon2):
    # The math module contains a function named
    # radians which converts from degrees to radians.
    lon1 = radians(lon1)
    lon2 = radians(lon2)
    lat1 = radians(lat1)
    lat2 = radians(lat2)
    # Haversine formula
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * asin(sqrt(a))
    # Radius of earth in kilometers. Use 3956 for miles
    r = 6371
    # calculate the result
    return(c * r)
```

```python
dist = {}
dist["ATL"] = 0
dist["FTY"] = int(cal_distance(-84.4418, 33.6301, -84.52, 33.78))
dist["HMP"] = int(cal_distance(-84.4418, 33.6301, -84.331, 33.3899))
dist["PDK"] = int(cal_distance(-84.4418, 33.6301, -84.302, 33.8756))
dist
```

```python
### Chosen the 3 met stations closest to Atlanta airport - FTY, ATL, PDK

atlanta_closest_data = pd.read_csv("asus_closest.csv", names=['station','valid','lon','lat','tmpf','dwpf','relh','drct','sknt','p01i','alti','mslp','vsby','gust','skyc1','skyc2','skyc3','skyc4','skyl1','skyl2','skyl3','skyl4','wxcodes','ice_accretion_1hr','ice_accretion_3hr','ice_accretion_6hr','peak_wind_gust','peak_wind_drct','peak_wind_time','feel','metar','snowdepth']
                                   , skiprows=1, sep=',')

atlanta_closest_data.head()
```

```python
import numpy as np

atlanta_closest_data.head()
calc_indices = [5, 6, 7, 8, 9, 10, 11, 11, 12, 13, 18, 19, 20, 21, 23, 24, 25, 26, 27]

new_df = pd.DataFrame()

i = 0

while(i<len(atlanta_closest_data)//2):
    line = atlanta_closest_data.loc[i, :]
    aggr_line = np.empty((0, len(line)))
    aggr_line = np.array([line])
    
    #Extracting the hour
    hour = line["valid"].split(" ")[1].split(":")[0]
    
    k = i+1
    while(k < len(atlanta_closest_data)//2 and atlanta_closest_data.loc[k,:]["valid"].split(" ")[1].split(":")[0] == hour):
        temp = np.array([atlanta_closest_data.loc[k]])
        aggr_line = np.append(aggr_line, temp, axis = 0)
        k += 1
        
    i = k
    for j in range(len(calc_indices)):
        
        tot = 0
        div = 0
        for m in range(len(aggr_line)):
            if((type(aggr_line[m][calc_indices[j]]) == type('') and not aggr_line[m][calc_indices[j]]) or (type(aggr_line[m][calc_indices[j]]) != type('') and not np.isnan(aggr_line[m][calc_indices[j]]))):
                tot += (dist[aggr_line[m][0]] * float(aggr_line[m][calc_indices[j]]))
                div += dist[aggr_line[m][0]]
        
        if(div != 0):
            line[calc_indices[j]] = tot/div
        else:
            line[calc_indices[j]] = np.nan
            
    new_df = new_df.append(line)
    
i += 1
        
    
new_df.head()
```
